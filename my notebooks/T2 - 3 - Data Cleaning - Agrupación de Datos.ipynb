# Agregación de Datos por categorías

```python
import numpy as np
import pandas as pd
np.random.seed(2019)
```

```python
#categorias de trabajo
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
#parametros para crear data set dummy

n = 500
gender_data = []
income_data = []

for i in range(0, n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[1:10]
```

```python
income_data[1:10]
```

```python
#crear otros parametros para agregar al data set

height = 160 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 20 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
#armar el data set usando la función DataFrame() de pandas

data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data, 
        "Height" : height, 
        "Weight" : weight, 
        "Age" : age, 
        "Income" : income
    }
)
```

```python
data.head()
```

```python
# Agrupar datos 

grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

## Operaciones sobre datos agrupados

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
grouped_income = double_group["Income"]
```

```python
grouped_income.describe()
```

```python
double_group.aggregate(
    {
        "Income" : np.sum,
        "Age" : np.mean,
        "Height" : np.std
    }
)
```

```python
double_group.aggregate({
    "Age" : np.mean,
    "Height" : lambda h:(np.mean(h))/np.std(h) #lambda es el valor tipico, o tificación.
})
```

```python
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```

```python
double_group.aggregate([np.sum, np.mean, np.std])
```

## Filtrado de Datos

```python
double_group["Age"].filter(lambda x: x.sum() > 2400)
```

## Transformación de variables

```python
zscore = lambda x : (x - x.mean()/x.std())
```

```python
z_group = double_group.transform(zscore).head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(z_group["Age"])
```
