## Simulación de Monte Carlo

* Generar 2 numeros aleatorios entre 0 y 1, x e y
* Calcular x^2 + y^2
    * Si el valor es inferior a 1 estoy dentro del circulo
    * Si es superior a 1, estamos fuera del circulo

* Calcular el número total de veces que estan dentro del círculo y lo dividimos entre el número total de intentos para obtener una aproximaación de la probabilidad de caer dentro del círculo.
* Usamos dicha probabilidad para aproximar el valor de π.
* Repetimos el experimento un número de vecs (p ej 1000), para obtener (1000) diferentes aproximaciones de π.
* Calculamos el promedio de los 1000 experimientos anteriores para dar un valor final a π.

    

```python
pi_avg = 0
n = 1000
pi_value_list = []
n_exp = 100
for i in range(n_exp):
    value = 0
    x = np.random.uniform(0,1,n).tolist()
    y = np.random.uniform(0,1,n).tolist()
    for j in range(n):
        z = np.sqrt(x[j] * x[j] + y[j] * y[j])
        if z <= 1:
            value += 1
    float_value = float(value)
    pi_value = float_value * 4 / n
    pi_value_list.append(pi_value)
    pi_avg += pi_value
    
pi = pi_avg / n_exp
print(pi)
plt.plot(pi_value_list)
```
