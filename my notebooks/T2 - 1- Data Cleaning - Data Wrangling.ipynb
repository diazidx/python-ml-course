## Generar Numeros aleatorios

```python
import numpy as np
```

```python
#generar un número aleatorio enterio entre 1 y 100
np.random.randint(1, 100)
```

```python
#generar un numero aleatorio entre 0 y 1
np.random.random()
```

```python
#Función que genera una lista de numeros aleatorios enteros dentro del intervalo [a, b] 

def randint_list(n, a, b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a, b))
    return x
```

```python
randint_list(20, 5, 35)
```

```python
# otra alternativa para agregar numeros random
import random as rnd
```

```python
for i in range(10):
    print(rnd.randrange(0, 100, 7))
```
